```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df.drop(columns = ['User ID'],inplace=True)
```

```python
df
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='Purchased'),df.Purchased,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_train.isnull().sum()
```

```python
trf1 = ColumnTransformer([
    ('ohe_gender',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[0])
],remainder='passthrough')
```

```python
trf2 = ColumnTransformer([
    ('scaler',StandardScaler(),[1,2])
])
```

```python
trf3 = DecisionTreeClassifier()
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred,y_test)*100
score
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
